## We have 4 dimension table and 1 fact table , so now i need to create a Star Schema 

```python
df_join1 = df_fact.join(df_skill_category.set_index('skill_category_id'), on='skill_category_id', how='left')
```

```python
display(df_join1)
```

```python
df_join2 = df_join1.join(df_industry.set_index('industry_id'), on='industry_id', how='left')
```

```python
display(df_join2)
```

```python
df_join3 = df_join2.join(df_date.set_index('date_id'), on='date_id', how='left')
display(df_join3)
```

```python
df_final = df_join3.join(df_dim_country.set_index('country_id'), on='country_id', how='left')
display(df_final)
```

```python
df_final.info()
```

## Need to do some filter on column and make the data much more useful for use 


```python
df_final = df_final.drop(columns=['index_id'])
df_final = df_final.drop(columns=['date_id'])
df_final = df_final.drop(columns=['skill_category_id'])
df_final = df_final.drop(columns=['industry_id'])
df_final = df_final.drop(columns=['country_id'])
display(df_final)
df_final.info()
```
